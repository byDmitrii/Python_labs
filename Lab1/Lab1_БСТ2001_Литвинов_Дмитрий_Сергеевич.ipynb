### Задание №3

Реализовать методы сортировки строк числовой матрицы в соответствии с заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со временем стандартной функции сортировки. Испытания проводить на сгенерированных матрицах.

Методы: Выбором | Вставкой | Обменом | Шелла | Турнирная | Быстрая сортировка | Пирамидальная

```python
#сортировка выбором
import time

sel_start_time = time.time()

def selection_sort(Two_Dimen_Matrix):
    mas=Two_Dimen_Matrix.copy()
    for i in range(len(mas)-1):
        m = i
        j = i + 1
        while j < len(mas):
            if mas[j] < mas[m]:
                m = j
            j = j + 1
        mas[i], mas[m] = mas[m], mas[i]
    return(mas)
for ind in range(n):
    print(selection_sort(Two_Dimen_Matrix[ind]))

print(f"{(time.time() - sel_start_time)*1000} миллисекунд")
```

```python
#сортировка вставкой

insert_start_time = time.time()

def insertion_sort(Two_Dimen_Matrix):
    mas=Two_Dimen_Matrix.copy()
    for i in range(1, len(mas)):
        item_to_insert = mas[i]
        j = i - 1
        while j >= 0 and item_to_insert < mas[j]:
            mas[j+1] = mas[j]
            j -= 1
        mas[j+1] = item_to_insert
    return(mas)
for ind in range(n):
    print(insertion_sort(Two_Dimen_Matrix[ind]))

print(f"{(time.time() - insert_start_time)*1000} миллисекунд")
```

```python
#сортировка обменом

bubble_start_time = time.time()

def bubble_sort(Two_Dimen_Matrix):
    mas=Two_Dimen_Matrix.copy()
    for i in range(len(mas)):
        for j in range(len(mas)-i-1):
            if mas[j] > mas[j+1]:
                mas[j], mas[j+1] = mas[j+1], mas[j]
    return(mas)
for ind in range(n):
    print(bubble_sort(Two_Dimen_Matrix[ind]))
    
print(f"{(time.time() - bubble_start_time)*1000} миллисекунд")
```

```python
#сортировка Шелла

Shell_start_time = time.time()

def Shell_sort(Two_Dimen_Matrix):
    mas=Two_Dimen_Matrix.copy()
    interval = len(mas)//2
    while interval:
        for i, el in enumerate(mas):
            while i >= interval and mas[i - interval] > el:
                mas[i] = mas[i - interval]
                i -= interval
                mas[i] = el
        interval = 1 if interval == 2 else int(interval * 5.0 / 11)
    return(mas)

for ind in range(n):
    print(Shell_sort(Two_Dimen_Matrix[ind]))
    
print(f"{(time.time() - Shell_start_time)*1000} миллисекунд")
```

```python
#Турнирная сортировка

tournament_start_time = time.time()

def tournamentSort(Two_Dimen_Matrix):
    mas = Two_Dimen_Matrix.copy()
    tree = [None] * 2 * (len(mas) + len(mas) % 2)
    index = len(tree) - len(mas) - len(mas) % 2
    
    for i, v in enumerate(mas):
        tree[index + i] = (i, v)
    
    for j in range(len(mas)):
        n = len(mas)
        index = len(tree) - len(mas) - len(mas) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1) 
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        mas[j] = x
        tree[len(tree) - len(mas) - len(mas) % 2 + index] = None
    return(mas)
for ind in range(n):
    print(tournamentSort(Two_Dimen_Matrix[ind]))

print(f"{(time.time() - tournament_start_time)*1000} миллисекунд")
```

```python
#Быстрая сортировка

quick_start_time = time.time()

def quick_sort(Two_Dimen_Matrix):
    mas = Two_Dimen_Matrix.copy()
    if len(mas) < 2:
        return mas
    low, same, high = [], [], []
    pivot = mas[random.randint(0, len(mas) - 1)]
    for item in mas:
        if item < pivot:
            low.append(item)
        elif item == pivot:
            same.append(item)
        elif item > pivot:
            high.append(item)
    return quick_sort(low) + same + quick_sort(high)

for ind in range(n):
    print(quick_sort(Two_Dimen_Matrix[ind]))

print(f"{(time.time() - quick_start_time)*1000} миллисекунд")
```

```python
#пирамидальная сортировка

pyramid_start_time = time.time()

def heapify(array, heapSize, rootIndex):
    largest = rootIndex
    leftChild = (2 * rootIndex) + 1
    rightChild = (2 * rootIndex) + 2
    if leftChild < heapSize and array[leftChild] > array[largest]:
        largest = leftChild
    if rightChild < heapSize and array[rightChild] > array[largest]:
        largest = rightChild
    if largest != rootIndex:
        array[rootIndex], array[largest] = array[largest], array[rootIndex]
        
def heap_sort(matrix):
    array = matrix.copy()
    n = len(array)
    for i in range(n, -1, -1):
        heapify(array, n, i)
    for i in range(n - 1, 0, -1):
        array[i], array[0] = array[0], array[i]
        heapify(array, i, 0)
    return(array)

for ind in range(n):
    print(heap_sort(Two_Dimen_Matrix[ind]))

print(f"{(time.time() - pyramid_start_time)*1000} миллисекунд")
```

```python
#встроенная сортировка

integral_start_time = time.time()

def integral_sort(matrix):
    mas = matrix.copy()
    mas.sort()
    return(mas)
    
for ind in range(n):
    print(integral_sort(Two_Dimen_Matrix[ind]))
    
print(f"{(time.time() - integral_start_time)*1000} миллисекунд")
```
